```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url_list = []
prices_list = []
propTitles = []
propValues = []

for i in range(1,21):
    url = "https://www.arabam.com/ikinci-el/otomobil?take=50&page=" + str(i)
    r = requests.get(url)
    source = BeautifulSoup(r.content,"lxml")
    
    urls = source.find_all("div",attrs={"class":"pr10 fade-out-content-wrapper"})
    for url in urls:
        url_car = "https://www.arabam.com"+url.a.get("href")
        url_list.append(url_car)
        print(url_car)
        
    r_car = requests.get(url_car)
    source_car = BeautifulSoup(r_car.content,"lxml")
        
    properties = source_car.find_all("div",attrs={"class":"property-item"})
    for prop in properties:
        prop_title = prop.find("div",attrs={"class":"property-key"}).text.replace("\n", "").replace("\r", "").strip()
        prop_value = prop.find("div",attrs={"class":"property-value"}).text.replace("\n", "").replace("\r", "").replace("Kopyaland", "").strip()
        propTitles.append(prop_title)
        propValues.append(prop_value)
        
    prices = source.find_all("span",attrs={"class":"db no-wrap listing-price"})
    for price in prices:
        prices_list.append(price.text.replace("\r\n", "").strip())
        print(price.text.replace("\r\n", "").strip())

print(str(len(url_list))+" adet link bulundu.")
print(str(len(prices_list))+" adet fiyat bulundu.")
print(str(len(propTitles))+" adet zellik bal bulundu.")
print(str(len(propValues))+" adet zellik verisi bulundu.")
```

```python
df_urls = pd.DataFrame()
df_urls["Url"] = url_list
df_urls["Fiyat"] = prices_list
```
