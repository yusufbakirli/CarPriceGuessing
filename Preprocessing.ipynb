```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("C:/Users/Yusuf/Desktop/proje/machine learning/car_data.xlsx")
```

```python
df = df.drop(['Kimden','Model','Motor Hacmi','Ort. Yakıt Tüketimi','Renk','Takasa Uygun','Yakıt Deposu','Çekiş','İlan No','İlan Tarihi','Url'], axis=1)
```

```python
df = df.rename(columns={'Boya-değişen': 'Boya_Degisen','Kasa Tipi': 'Kasa_Tipi','Motor Gücü': 'Motor_Gucu',
                        'Seri': 'Model','Vites Tipi': 'Vites_Tipi','Yakıt Tipi': 'Yakit_Tipi',
                        'Yıl': 'Yil'})
```

```python
df.columns
```

```python
df.head()
```

```python
df["Fiyat"] = df.Fiyat.str.replace('.', '')
df["Fiyat"] = df.Fiyat.str.replace(' TL', '')
df["Fiyat"] = df.Fiyat.astype(float)
```

```python
df
```

```python
df["Boya_Degisen"] = df.Boya_Degisen.str.replace('Tamamı orjinal','0')
df["Boya_Degisen"] = df.Boya_Degisen.str.replace('Tamamı boyalı','14')
df["Boya_Degisen"] = df.Boya_Degisen.str.replace('Belirtilmemiş','0')
df["Boya_Degisen"] = df.Boya_Degisen.str.replace(' değişen','')
df["Boya_Degisen"] = df.Boya_Degisen.str.replace(' boyalı','')
df["Boya_Degisen"] = df.Boya_Degisen.str.replace(' ','')
def add_numbers(cell):
    numbers = [int(num) for num in cell.split(',')]
    return sum(numbers)
df["Boya_Degisen"] = df["Boya_Degisen"].apply(add_numbers)
df['Kasa_Tipi'] = df['Kasa_Tipi'].replace(['Hatchback/5','Hatchback/3'],'Hatchback')
df["Kilometre"] = df.Kilometre.str.replace('.', '')
df["Kilometre"] = df.Kilometre.str.replace(' km', '')
df["Motor_Gucu"] = df.Motor_Gucu.str.replace(' hp', '')
mode_motorgucu = df["Motor_Gucu"].mode()[0]
df["Motor_Gucu"].fillna(mode_motorgucu, inplace=True)
```

```python
df
```

```python
df["Boya_Degisen"] = df.Boya_Degisen.astype(float)
df["Kilometre"] = df.Kilometre.astype(float)
df["Motor_Gucu"] = df.Motor_Gucu.astype(float)
df["Yil"] = df.Yil.astype(int)
```
